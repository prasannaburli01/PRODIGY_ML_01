```python
# House Price Prediction using Linear Regression
# Prodigy Infotech - ML Task 01

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Create simple dataset
data = {
    'Area': [800, 900, 1000, 1100, 1200, 1300, 1400, 1500],
    'Bedrooms': [1, 2, 2, 3, 3, 3, 4, 4],
    'Bathrooms': [1, 1, 2, 2, 2, 3, 3, 3],
    'Price': [50000, 60000, 75000, 85000, 95000, 110000, 130000, 150000]
}

df = pd.DataFrame(data)

# Features and target
X = df[['Area', 'Bedrooms', 'Bathrooms']]
y = df['Price']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Prediction
prediction = model.predict([[1200, 3, 2]])

print("Predicted House Price:", prediction[0])
print("Task 01 completed successfully")
```
